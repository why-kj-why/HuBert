```python
def normalize(df):
    """
        Uses minMax scaler on data to normalize. Important especially for Volume and google_lvl
        @param df: data frame with all features
    """
    df = DataFrame(df)
    df.dropna(inplace = True)
    df = df.values
    df = df.astype('float32')
    scaler = MinMaxScaler(feature_range=(-1, 1))
    norm = scaler.fit_transform(df)
    return scaler, norm


scaler1, normalized_data1 = normalize(np.transpose(VO_imfs))
scaler2, normalized_data2 = normalize(np.transpose(VH_imfs))
scaler3, normalized_data3 = normalize(np.transpose(VL_imfs))
scaler4, normalized_data4 = normalize(np.transpose(VC_imfs))

MEGA = np.stack([normalized_data1, normalized_data2, normalized_data3, normalized_data4], axis=-1)

train_size = int(len(MEGA) * 0.8)
test_size = len(MEGA) - train_size
train, test = MEGA[:train_size], MEGA[train_size:]

def create_dataset(dataset, lookback):
    X, y = [], []
    for i in range(len(dataset)-lookback):
        feature = dataset[i:i+lookback]
        target = dataset[i+1:i+lookback+1]
        X.append(feature)
        y.append(target)
    return torch.tensor(X), torch.tensor(y)

lookback = 1
X_train, y_train = create_dataset(train, lookback=lookback)
X_test, y_test = create_dataset(test, lookback=lookback)

class lstmModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.lstm1 = nn.LSTM(input_size=6, hidden_size=100, num_layers=1, batch_first=True)
        self.linear1 = nn.Linear(100, 6)
        self.lstm2 = nn.LSTM(input_size=6, hidden_size=200, num_layers=1, batch_first=True)
        self.linear2 = nn.Linear(200, 6)
    def forward(self, x):
        x, _ = self.lstm1(x)
        x = self.linear1(x)
        x, _ = self.lstm2(x)
        x = self.linear2(x)
        return x
```

```python
model0 = lstmModel()
optimizer = optim.Adam(model0.parameters())
loss_fn = nn.MSELoss()
loader = data.DataLoader(data.TensorDataset(X_train[:, :, :, 0], y_train[:, :, :, 0]), shuffle=True, batch_size=8)

n_epochs = 30
for epoch in range(n_epochs):
    model0.train()
    for X_batch, y_batch in loader:
        y_pred = model0(X_batch)
        loss = loss_fn(y_pred, y_batch)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    # Validation
    if epoch % 10 != 0:
        continue
    model0.eval()
    with torch.no_grad():
        y_pred = model0(X_train[:, :, :, 1])
        train_rmse = np.sqrt(loss_fn(y_pred, y_train[:, :, :, 0]))
        y_pred = model0(X_test[:, :, :, 0])
        test_rmse = np.sqrt(loss_fn(y_pred, y_test[:, :, :, 0]))
    print("Epoch %d: train RMSE %.4f, test RMSE %.4f" % (epoch, train_rmse, test_rmse))

with torch.no_grad():
    # shift train predictions for plotting
    train_plot = np.ones_like(MEGA[:, :, 0]) * np.nan
    y_pred = model0(X_train[:, :, :, 0])
    y_pred = y_pred[:, -1, :]
    train_plot[lookback:train_size] = model0(X_train[:, :, :, 0])[:, -1, :]
    # shift test predictions for plotting
    test_plot = np.ones_like(MEGA[:, :, 0]) * np.nan
    test_plot[train_size+lookback:len(MEGA[:, :, 0])] = model0(X_test[:, :, :, 0])[:, -1, :]
    
    
    
plt.subplot(3, 1, 1)
plt.plot(MEGA[:, :, 0])
plt.subplot(3, 1, 2)
plt.plot(train_plot, c='r')
plt.subplot(3, 1, 3)
plt.plot(test_plot, c='g')
plt.show()
```
